# Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])

print("==== Sales Analysis ====")
print("Sales data shape: ", sales_data.shape)
print("Sample data for first 3 restaurant: \n", sales_data[:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))
year_total = np.sum(sales_data[:, 1:], axis=0)
print(year_total)

#Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

#Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

#Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
#Cumulative Sum 
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales across all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 0, 10])

print("Vector Addition: ", vector1 + vector2)
print("\nMultiplication vector: ", vector1 * vector2)
print("\nDot Product: ", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2)/ (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
